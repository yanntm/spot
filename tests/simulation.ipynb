```python
def is_equiv(is_equiv_aut1, is_equiv_aut2):
    if spot.complement(is_equiv_aut1).intersects(is_equiv_aut2):
        return False
    if spot.complement(is_equiv_aut2).intersects(is_equiv_aut1):
        return False
    return True

def test(test_aut):
    test_res = spot.reduce_direct_sim(test_aut)
    test_ref = spot.simulation(test_aut)

    print('States: ' +  str(test_aut.num_states()) \
        + ' -> ' + str(test_res.num_states()) \
        + ' (simulation: ' + str(test_ref.num_states()) + ')')

    print('Edges: ' +  str(test_aut.num_edges())
        + ' -> ' + str(test_res.num_edges())
        + ' (simulation: ' + str(test_ref.num_edges()) + ')')
    
    return is_equiv(test_aut, test_res)
```

```python
def print_sim(print_sim_aut):
    a = spot.find_direct_simulation(print_sim_aut)

    print(' ', end=' ')
    for i in range(0, print_sim_aut.num_states()):
        print(i, end=' ')
    print()
    
    for i in range(0, print_sim_aut.num_states()):
        print(i, end=' ')
        for j in range(0, print_sim_aut.num_states()):
            if a[i * print_sim_aut.num_states() + j]:
                print('x', end=' ')
            else:
                print('.', end=' ')
        print()
```

```python
fail = 0
fail_aut = []
for a in spot.automata('randaut -H -n50 4|'):
    if not test(a):
        fail += 1
        fail_aut.append(a)
    print()
assert fail == 0
```
